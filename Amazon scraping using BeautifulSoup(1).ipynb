```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL='https://www.amazon.in/s?k=phone&crid=36MZNU59ONBPT&sprefix=phone%2Caps%2C227&ref=nb_sb_noss_1'
```

```python
#Headers for request
HEADERS=({'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0','Accept-Language': 'en-US,en;g=0.5'})
```

```python
#http request
webpage=requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
#soup object containing all data
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
soup
```

```python
#Fetch links as List of Tag Objects
links=soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```

```python
link=links[0].get('href')
```

```python
product_list="https://www.amazon.in"+link
```

```python
product_list
```

```python
#http request
new_webpage=requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
#soup object containing all data
new_soup=BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup.find("span",attrs={"id":'productTitle'}).text.strip()
```

```python
new_soup.find("span",attrs={"class":'a-price-whole'}).text
```
